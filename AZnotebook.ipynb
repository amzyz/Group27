# Sidewalk to Street Ratios (Walkability)

## By: Amy Zhang

##### Dataset: [Sidewalk Street Ratio](https://data.wprdc.org/dataset/sidewalk-to-street-walkability-ratio/resource/b90ccee1-c0aa-43b9-93e2-8a25e690c393?inner_span=True) and [Neighborhoods](https://data.wprdc.org/dataset/neighborhoods2/resource/668d7238-cfd2-492e-b397-51a6e74182ff?inner_span=True)

```python
# Load in walkability dataset
sidewalk_street_ratios = pd.read_csv("SidewalkStreeRatio.csv")
sidewalk_street_ratios.head()

# narrow down to relevant columns
walkability= pd.DataFrame(columns= ["GEOIDs","Ratios"])
walkability["GEOIDs"]= sidewalk_street_ratios["GEOID"]
walkability["Ratios"]= sidewalk_street_ratios["Ratio"]
walkability.head(3)

# I only want the ratios that are above 1 (more room for scurrying)
wideWalksOnly= walkability[walkability["Ratios"]>1.0]
wideWalksOnly.head(10)

# Sort by top values (descending, highest values at top)
wideWalksOnly.sort_values("Ratios", ascending = False)

# make a new dataframe with neighborhood names as an empty column
Neighborhood_Ratios= wideWalksOnly.DataFrame(columns= [""])
```
