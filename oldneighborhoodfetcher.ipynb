```python
# import module
import pandas as pd
import csv
from geopy.geocoders import Nominatim

# initialize coords from csv
places = pd.read_csv("alleghenyassetmap.csv", index_col="_id", dtype='string')

# initialize Nominatim API
geolocator = Nominatim(user_agent="JupyterLab")
 
for index, row in places.iterrows():
    # Latitude & Longitude input
    Latitude = row['latitude']
    Longitude = row['longitude']

    location = geolocator.reverse(Latitude+","+Longitude)

    address = location.raw['address']

    # traverse the data
    neighbourhood = address.get('neighbourhood')
    town = address.get('town')
    village = address.get('village')
    city = address.get('city', '')
    state = address.get('state', '')
    country = address.get('country', '')
    code = address.get('country_code')
    zipcode = address.get('postcode')
    #print('Address : ', address)
    
    if (neighbourhood is not None):
        localarea = 'Neighbourhood', neighbourhood
    elif(town is not None):
        localarea = 'Town', town
    elif(village is not None):
        localarea = 'Village', village
    else:
        localarea = 'City', city
    print(localarea)
    
    
    #print('Neighbourhood : ', neighbourhood)
    #print('City : ', city)
    #print('State : ', state)
    #print('Country : ', country)
    #print('Zip Code : ', zipcode)
```
